```python
#SinglyLinkedList implementation

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __str__(self):
        current = self.head
        result = []
        while current:
            result.append(str(current.data))
            current = current.next
        return " -> ".join(result)

    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            count = 0
            current = self.head
            prev = None
            if index == count:
                old_head = self.head
                self.head = new_node
                new_node.next = old_head
            else:
                while current.next and count != index:
                    prev = current
                    current = current.next
                    count += 1
                if count == index:
                    prev.next = new_node
                    new_node.next = current
                else:
                    current.next = new_node

    def remove(self, value):
        current = self.head
        previous = None
        while current:
            if current.data == value:
                if previous:
                    previous.next = current.next
                else:
                    self.head = current.next
                return
            previous = current
            current = current.next

    def index(self, index, value):
        current = self.head
        count = 0
        while current:
            if count == index:
                return current.data
            current = current.next
            count += 1
        return None
```

```python
# Create a new singly linked list
l = SinglyLinkedList()
l.append(10)
l.append(20)
l.append(30)

print("List after appending numbers:")
print(l)  # Output: 10 -> 20 -> 30
l.remove(20)
print(l)  # Output: 10 -> 30
```

```python
#Doubly Linked Lists

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            self.prev = None

    def __str__(self):
        result = []
        current = self.head
        while current:
            result.append(str(current.data))
            current = current.next
        return " <-> ".join(result)

    
    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
    
    def index(self, index):
        current = self.head
        count = 0
        while current:
            if count == index:
                return current.data
            current = current.next
            count += 1
        return None

    def insert(self, index, value):
        new_node = self.__Node(value)
        if index == 0:
            if self.head:
                self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            current_index = 0
            while current and current_index < index:
                current = current.next
                current_index += 1
            if current:
                new_node.next = current
                new_node.prev = current.prev
                if current.prev:
                    current.prev.next = new_node
                current.prev = new_node
            else:
                self.append(value)

    def remove(self, value):
        current = self.head
        while current:
            if current.data == value:
                if current.prev:
                    current.prev.next = current.next
                if current.next:
                    current.next.prev = current.prev
                if current == self.head:  # If it's the head element
                    self.head = current.next
                return
            current = current.next



```

```python
dl = DoublyLinkedList()
dl.append(10)
dl.append(20)
dl.append(30)

print("List after appending numbers:")
print(dl)  

dl.insert(1, 15)
print("After Inserting 15 at index 1:", dl)  # Output: After Inserting 15 at index 1: 10 <-> 15 <-> 20 <-> 30

dl.remove(20)
print("After Removing 20:", dl)  # Output: After Removing 20: 10 <-> 15 <-> 30

print("Value at index 2:", dl.index(2))  # Output: Value at index 2: 30
```
